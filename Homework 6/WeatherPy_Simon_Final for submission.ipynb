```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time


# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
# Temperature (F) vs. Latitude
# Humidity (%) vs. Latitude
# Cloudiness (%) vs. Latitude
# Wind Speed (mph) vs. Latitude

# for gathering list of unique cities
lat_lngs = []
cities = []

# for generating lists of values from the calls
city_name=[]
Latitude = []
Temperature = []
Humidity = []
Cloudiness = []
Wind_Speed = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)


# Print the city count to confirm sufficient count
len(cities)
```

```python

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 
city_count=0
for city in cities:
    time.sleep(1)
    city_count=city_count + 1
    complete_call_specs = url + "&q=" + city  
    response = requests.get(complete_call_specs)
    Translated_response = response.json()
    try:
        response_lat =Translated_response ['coord'] ['lat']
        response_temperature = Translated_response ['main']['temp']
        response_humidity = Translated_response ['main']['humidity']
        response_cloudiness = Translated_response ['clouds']['all']
        response_wind_speed = Translated_response ['wind']['speed']
        city_name.append(city)
        Latitude.append(response_lat)
        Temperature.append(response_temperature)
        Humidity.append(response_humidity)
        Cloudiness.append(response_cloudiness)
        Wind_Speed.append(response_wind_speed)
        print("City #"+str(city_count)+ "   City Name: " + city)
    except:
        city_count=city_count-1
    
```

```python
#build dataframe
full_city_dataset = pd.DataFrame()
full_city_dataset['Latitude']  = Latitude
full_city_dataset['Temperature']  = Temperature
full_city_dataset['Humidity']  = Humidity
full_city_dataset['Cloudiness']  = Cloudiness
full_city_dataset['Wind_Speed']  = Wind_Speed

```
